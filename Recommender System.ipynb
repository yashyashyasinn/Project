```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
from scipy.spatial import distance
```

# Pearson Correlation

```python
data = pd.read_excel('RecSys.xlsx')
data
```

```python
hitungPearson=[]
hitungPearson = data
hitungPearson['sim_pearson'] = 1

x = np.array(data.iloc[0,1:9])

for i in range (1,len(hitungPearson)):
  y = np.array(data.iloc[i,1:9])
  sim = scipy.stats.pearsonr(x, y)
  sim = sim[0]
  hitungPearson.iloc[i:,10]=sim
```

```python
#menghilangkan baris dengan nilai similarity dibawah 0.5
hitungPearson.drop(hitungPearson[hitungPearson.sim_pearson <0.5].index, inplace=True)
hitungPearson
```

```python
#hasil prediksi rating alesya berdasarkan pearson correlation
a = len(hitungPearson)
pred_rat_alesya_com3 = sum(hitungPearson.iloc[1:a,9]) / (a-1)
pred_rat_alesya_com3
```

# Euclidean Similarity

```python
data = pd.read_excel('RecSys.xlsx')

hitungEu=[]
hitungEu = data
hitungEu['sim_cos'] = 1

x = np.array(data.iloc[0,1:9])

for i in range (1,len(hitungEu)):
  y = np.array(data.iloc[i,1:9])
  sim = np.linalg.norm(x-y)
  hitungEu.iloc[i:,10]=sim
```

```python
#menghilangkan baris dengan nilai similarity diatas 4
hitungEu.drop(hitungEu[hitungEu.sim_cos > 4].index, inplace=True)
hitungEu
```

```python
#prediksi rating alesya berdasarkan euclidian
a = len(hitungEu)
pred_cos_rat_alesya_com3 = sum(hitungEu.iloc[1:a,9]) / (a-1)
pred_cos_rat_alesya_com3
```
